```python
import json
import utils

import aisuite as ai
from dotenv import load_dotenv

load_dotenv()

# Define the client. You can use this variable inside your graded functions!
CLIENT = ai.Client()
```

```python
def generate_draft(topic: str, model: str = "openai:gpt-4.1") -> str:
    ### START CODE HERE ###

    # Define your prompt here. A multi-line f-string is typically used for this.
    prompt = f"""
    You are a helpful assistant that generates a draft for a research paper. 
    
    The topic of the paper is "{topic}". 
    
    Please write a well-structured draft that includes an introduction, main points, and a conclusion.
    """ 

    ### END CODE HERE ###
    
    # Get a response from the LLM by creating a chat with the client.
    response = CLIENT.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=1.0,
    )

    return response.choices[0].message.content
```

```python
def reflect_on_draft(draft: str, model: str = "openai:o4-mini") -> str:

    ### START CODE HERE ###

    # Define your prompt here. A multi-line f-string is typically used for this.
    prompt = f"""
    You are a helpful assistant that provides feedback on research drafts.
    
    The draft is as follows:
    {draft}

    Requirements for feedback:
    1. The feedback should be critical but constructive.
    2. It should address issues such as structure, clarity, strength of argument, and writing style.
    3. The feedback should be actionable, providing specific suggestions for improvement.
    
    Return constructive feedback as string on the draft.
    """ 

    ### END CODE HERE ###

    # Get a response from the LLM by creating a chat with the client.
    response = CLIENT.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=1.0,
    )

    return response.choices[0].message.content
```

```python
def revise_draft(original_draft: str, reflection: str, model: str = "openai:gpt-4o") -> str:

    ### START CODE HERE ###

    # Define your prompt here. A multi-line f-string is typically used for this.
    prompt = f"""
You are a helpful assistant that revises research drafts based on feedback.

The original draft is as follows:
{original_draft}

The feedback on the draft is as follows:
{reflection}

Return only the final revised essay.

Requirements for revision:
1.The revised draft should address the issues mentioned in the feedback.
2. It should improve clarity, coherence, argument strength, and overall flow.

"""

    # Get a response from the LLM by creating a chat with the client.
    response = CLIENT.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=1.0,
    )

    ### END CODE HERE ###

    return response.choices[0].message.content
```

```python
def run_workflow(
        topic: str, 
        model_generate: str = "openai:gpt-4.1", 
        model_reflect: str = "openai:o4-mini", 
        model_revise: str = "openai:gpt-4o"
) -> None:
    utils.print_html(topic, title="Research Topic")

    # Step 1: Generate a draft
    draft = generate_draft(topic, model=model_generate)
    utils.show_output("Step 1 – Draft", draft, background="#fff8dc", text_color="#333333")

    # Step 2: Reflect on the draft
    feedback = reflect_on_draft(draft, model=model_reflect)
    utils.show_output("Step 2 – Reflection", feedback, background="#e0f7fa", text_color="#222222")

    # Step 3: Revise the draft based on feedback
    revised = revise_draft(draft, feedback, model=model_revise)
    utils.show_output("Step 3 – Revision", revised, background="#f3e5f5", text_color="#222222")
```

```python
topic = "Should social media platforms be regulated by the government?"
model_generate = "openai:gpt-4.1"
model_reflect = "openai:o4-mini"
model_revise = "anthropic:claude-sonnet-4-6"
run_workflow(
    topic=topic,
    model_generate=model_generate,
    model_reflect=model_reflect,
    model_revise=model_revise
)
```
